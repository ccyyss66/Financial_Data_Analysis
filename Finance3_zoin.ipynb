```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
t=np.arange(0.0,2.0,0.01)#起点0，终点2，步长0.01
#根据关系构造s
s=[]   
for i in t:
    s.append(1+sin(2*pi*i))
#设置标签和其他属性
plt.xlabel('time(s)')
plt.ylabel('voltage(mv)')
plt.title('About as simple as it gets,folks')
plt.grid(True)
plt.plot(t, s, 'b-', linewidth=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
t=np.linspace(0.0,5.0,50)#线性分割区间，第一个参数为起点位置，第二个参数为终点位置，第三个参数为分割的个数，默认50
#根据关系构造对应的因变量
y1=[]
y2=[]
for i in t:
    y1.append(cos(2*pi*i)*exp(-i))
    y2.append(cos(2*pi*i))
plt.subplot(2,1,1)#将画布分为2行1列，此时选择的是第一个子画布（第一个参数为画布行数，第二个参数为画布列数，第三个为选择的第几个画布）
#设置标签和属性
#本来想设中文标签和标题，然后默认的话中文是乱码的，得改设置才行。。。然后就用了英文
plt.ylabel('Damped oscillation')
plt.title('Damped And Undamped Oscillation')
plt.grid(True)
plt.plot(t,y1)
plt.plot(t,y1,'bo')

#选择第二个子画布并设置标签和属性
plt.subplot(2,1,2)
plt.xlabel('time (s)')
plt.ylabel('Undamped oscillation')
plt.grid(True)
plt.plot(t,y2)
plt.plot(t,y2,'bo')

plt.show()
```

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

mu, sigma = 100, 15
x = mu + sigma*np.random.randn(1000)#emmm.... 1000组数据还不够多，最好更大点

'''
直方图：
返回值：
    n: 直方图向量，是否归一化由参数normed设定
    bins: 返回各个bin的区间范围
    patches: 返回每个bin里面包含的数据，是一个list
参数：
    arr（必须）: 需要计算直方图的一维数组
    bins: 直方图的柱数，可选项，默认为10
    normed: 是否将得到的直方图向量归一化。默认为0,不归一的时候显示的是频度，归一的时候显示的是密度
    facecolor: 直方图颜色
    edgecolor: 直方图边框颜色
    alpha: 透明度
    histtype: 直方图类型，‘bar’, ‘barstacked’, ‘step’, ‘stepfilled’


'''
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='b', alpha=0.75)

# 添加虚化线
y = mlab.normpdf( bins, mu, sigma)  #画一条逼近的曲线  
l = plt.plot(bins, y, 'r--', linewidth=1)#选择线的一些参数

#设置一些标签
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  
from matplotlib import cm  
from matplotlib.ticker import LinearLocator, FormatStrFormatter  
import matplotlib.pyplot as plt  
import numpy as np  
  
fig = plt.figure()  
ax = fig.gca(projection='3d')  #3D类型
X = np.arange(-5, 5, 0.1)  #设置X,Y区间范围
Y = np.arange(-5, 5, 0.1)  
X, Y = np.meshgrid(X, Y)  
R = np.sqrt(X**2 + Y**2)  
Z = np.sin(R)
'''
入参：
X,Y,Z:坐标
rstride:横向步长
cstride：纵向步长
cmap：颜色为冷暖图
linewidth:线宽
antialiased：表面是否透明
'''
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)  

#坐标系的下边界和上边界  
ax.set_zlim(-1.01, 1.01)
#设置Z轴标度  
ax.zaxis.set_major_locator(LinearLocator(10))
#Z轴精度  
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
#shrink颜色条伸缩比例（0-1），aspect颜色条宽度（反比例，数值越大宽度越窄）  
fig.colorbar(surf, shrink=0.5, aspect=5)
#显示
plt.show()  
```
