* 习题一

```python
'''•	习题一
•	回想一下n！被读为“n阶乘”并定义为:
–	  n！= n×（n-1）×...×2×1
•	在各种模块中都有阶乘的计算，但是我们可以自己编写一个函数。
•	请写一个阶乘函数factorial，使得对于任何正整数n，函数factorial (n)返回n！
'''
def factorial(x):
    mul=1
    while x>0:
        mul*=x
        x=x-1
    return mul
factorial(5)
```

* 习题四

```python
'''	习题四
	考虑多项式：
	p(x)=a_0+a_1 x+a_2 x^2+⋯a_n x^n     （1）
	写一个函数p(x,coeff)，来计算公式计算（1），给定点x和一系列参数coeff
	尝试在循环中使用枚举enumerate() 
'''
def p(x,coeff):
    sum=0
    for index,item in enumerate(coeff):
        sum+=x**index*item
    return sum
p(2,[4,5,6])
```

```python
'''	习题五
	请你模拟和绘制相关的时间序列：
	 x_(t+1)=αx_t+ε_(t+1)       其中：x_0   =0  and   t=0,…,T
	 其中冲击序列{ε_t}假定为标准正态分布
	在你的解决方案中，限制你的导入语句：
	import numpy as np
	import matplotlib.pyplot as plt
	假设:T= 200, α=0.9
'''
import numpy as np
import matplotlib.pyplot as plt
T=200
alpha=0.9
eps=[np.random.randn() for i in range(T)]
x=[0]
for i in range(1,T+1):
    tmp=x[i-1]*alpha+eps[i-1]
    x.append(tmp)
#print(x)
plt.plot(x)
plt.show()

```

```python
'''现在从你的练习3开始，绘制三个模拟时间序列，每种情况分布为α= 0，α=0.8，α= 0.98'''
import numpy as np
import matplotlib.pyplot as plt
T=200
alpha=[0,0.8,0.98]
'''
[np.random.randn() for i in range(3*T)]
考虑到每个alpha对应的标准正态误差不一定相同
生成3*T个标准正态分布本来打算用上面这句创建总共3*T的标准正态分布，但考虑到0：T和T:2*T和2*T:3*T这三段之内并不一定是 独立同分布的，
所以生成了3个大小为T的标准正态再把他们拼接起来！
'''

e_a=[np.random.randn() for i in range(T)]
e_b=[np.random.randn() for i in range(T)]
e_c=[np.random.randn() for i in range(T)]
eps=[]
eps.append(e_a)
eps.append(e_b)
eps.append(e_c)
x=[[0],[0],[0]]
for i in range(1,4):
    for j in range(1,T+1):
        tmp=x[i-1][j-1]*alpha[i-1]+eps[i-1][j-1]
        x[i-1].append(tmp)
#print(x)
plt.plot(x[0], 'b-', label="alpha=0")
plt.plot(x[1], 'g-', label="alpha=0.8")
plt.plot(x[2], 'r-', label="alpha=0.98")
plt.legend()
plt.show()

```
