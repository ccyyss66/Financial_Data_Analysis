```python
import numpy as np
def p(x,coeff):
    tmp=np.ones(coeff.size)*x   #产生大小为n的值全为x的行向量
    mul_x=np.cumprod(tmp)/x     #生成x的i次方的行向量，除以x表示从x的0次方开始
    ele=coeff*mul_x             #数组中的对应元素相乘
    return sum(ele)            #返回累加和
coeff=np.array([1,2,3,4])       #定义系数对应的数组
x=2                         
ans=p(x,coeff)                  #调用
print(ans)
```

```python
import numpy as np
class Polynomial:                    #求多项式的类
    def __init__(self,coeff):        #初始化属性
        self.coeff=coeff
    def get_value(self,x):          #求多项式的值
        tmp=np.ones(coeff.size)*x    #产生大小为n的值全为x的行向量
        mul_x=np.cumprod(tmp)/x #   生成x的i次方的行向量，除x表示从x的0次方开始
        ele=coeff*mul_x             #数组中的对应元素相乘
        return sum(ele)            #返回累加和
    def diff(self):                #对多项式求导
        for (index,item) in enumerate(self.coeff):    #原来的系数乘以对应位置即为求导后的系数（下标从0开始）
            self.coeff[index]*=index
poly=Polynomial(np.array([1,2,3,4]))         #实例化对象，提供系数数组
val=poly.get_value(2)                        #求对应x的多项式值
poly.diff()                                  #对多项式求导
print(poly.coeff)
```

```python
from random import uniform
class ECDF:             #分布函数的类
    def __init__(self,samples):     #初始化样本
        self.observations=samples
    def __call__(self,x):           #求对应某一点的分布函数F（xi）的值
        sum=0                      
        for i in self.observations:
            if i<x:
                sum+=1               #统计出现次数
        return sum/len(self.observations)   #返回出现的概率
samples = [uniform(0, 1) for i in range(10)]   #生成0到1的随机数
F=ECDF(samples)                                 #实例化对象
ans=F(0.5)                             #调用
print(ans)
F.observations = [uniform(0, 1) for i in range(1000)]   #随机事件 n重伯努利试验 在实验次数够多的时候会接近一个定值
ans=F(0.5)
print(ans)
```

* 习题四


```python
import numpy as mp
import matplotlib.pyplot as plt

class ECDF:
    def __init__(self,x):     #初始化
        self.observations=x 
    def __call__(self,x):     #求对应某一点的分布函数F（xi）的值
        sum=0
        for i in self.observations:
            if i<x:
                sum+=1
        return sum/len(self.observations)   
    def calAndPlot(self,start,end):    #计算给定区间内点的F（xi）值，【区间内以默认50个点计】
        xi=np.linspace(start,end)      #未指定第三个参数默认取区间50个点
        f_xi=[]                        #分布函数
        for i in xi:                   #计算每个点
            f_xi.append(self.__call__(i))
        #print(f_xi)
        
        #画图
        plt.plot(xi,f_xi, 'b-', label="F(Xi)")
        plt.legend()
        plt.show()


x=np.random.randn(1000)    #随机数
F=ECDF(x)                  #实例化
F.calAndPlot(-4,4)         #调用画图
```
